# Pre-Processing

Importing libraries and loading data:

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv('data_EDA.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
numeric_df=df.select_dtypes(include=np.number)
```

```python
numeric_df.info()
```

```python
numeric_df.head()
```

```python
object_df=df.select_dtypes(exclude=np.number)
```

```python
object_df.info()
```

```python
object_df.head()
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(numeric_df)
names=numeric_df.columns
scaled_df = pd.DataFrame(scaled_df, columns=names) 
scaled_df.head()
```

```python
dummies=pd.get_dummies(object_df)
dummies.head()
```

```python
pp_df=pd.concat([scaled_df,dummies],axis=1)
```

```python
pp_df.head()
```

```python
pp_df.shape
```

```python
y=pp_df['SalePrice']
X=pp_df.loc[:, pp_df.columns!='SalePrice']
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=42, 
                                   test_size=0.3, 
                                   shuffle=True)
```

```python
print(X_train.head())
print(X_test.head())
print(y_train.head())
print(y_test.head())
```

```python
X_train.to_csv('X_train.csv', header=True, index=False)
X_test.to_csv('X_test.csv', header=True, index=False)
y_train.to_csv('y_train.csv', header=True, index=False)
y_test.to_csv('y_test.csv', header=True, index=False)
```
